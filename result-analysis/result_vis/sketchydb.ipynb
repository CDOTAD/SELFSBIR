```python
import os
```

```python
SKETCHYDB_ROOT = '/data1/zzl/dataset/sketchydatabase_official_split/'
```

```python
PHOTO_TEST = os.path.join(SKETCHYDB_ROOT, 'photo-test-all')
```

```python
PHOTO_TRAIN = os.path.join(SKETCHYDB_ROOT, 'photo-train-all')
```

```python
SKETCH_TEST = os.path.join(SKETCHYDB_ROOT, 'sketch-triplet-test-all')
```

```python
cnames = sorted(os.listdir(PHOTO_TEST))
```

```python
TEXT_TXT = '/home/bc/Work/Database/sketchy/info/testset.txt'
```

```python
NEW_ROOT = '/data1/zzl/dataset/sketchydb_fix/'
```

```python
FIX_PHOTO_TRAIN = os.path.join(NEW_ROOT, 'photo-train')
FIX_PHOTO_TEST = os.path.join(NEW_ROOT, 'photo-test')
FIX_SKETCH_TRAIN = os.path.join(NEW_ROOT, 'sketch-train')
FIX_SKETCH_TEST = os.path.join(NEW_ROOT, 'sketch-test')
```

```python
import tqdm
```

```python
cnt = 0
with open(TEXT_TXT) as test_file:
    lines = test_file.readlines()
    for line in tqdm.tqdm(lines):
        line = line.strip('\n')
        photo_src = os.path.join(PHOTO_TEST, line)
        cname, file_name = line.split('/')
        photo_class = os.path.join(FIX_PHOTO_TEST, cname)
        if not os.path.exists(photo_class):
            os.mkdir(photo_class)
        photo_dict = os.path.join(photo_class, file_name)
        shutil.copy(photo_src, photo_dict)
        p_name = file_name.split('.')[0]
        sketch_test_file = os.listdir(os.path.join(SKETCH_TEST, cname))
        for s_name in sketch_test_file:
            sp_name = s_name.split('.')[0]
            if sp_name.split('-')[0] == p_name:
                cnt += 1
                sketch_src = os.path.join(SKETCH_TEST, cname, s_name)
                sketch_class = os.path.join(FIX_SKETCH_TEST, cname)
                if not os.path.exists(sketch_class):
                    os.mkdir(sketch_class)
                sketch_dict = os.path.join(sketch_class, s_name)
                shutil.copy(sketch_src, sketch_dict)
    
```
